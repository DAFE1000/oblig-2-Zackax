```python
import numpy as np
import matplotlib.pyplot as plt

#x verdi fra ..
x = np.linspace(-0.5,3,100)


#definerer funksjonen
def f(x):
    return np.exp(-x/4)*np.arctan(x)
#derivert funksjon
def derivert(x):
    return np.exp(-x/4)*((-1/4)*np.arctan(x) + 1/((x**2)+1))
#dobbelt derivert funksjon
def andreDerivert(x):
    return np.exp(-x/4)*((1/16)*np.arctan(x)-(1/(2*(x**2+1)))-((2*x)/(x**2+1)**2))
#definerer variabel for grafene
fig,(ax1, ax2,ax3) = plt.subplots(3, 1, figsize=(7, 15))

#tanget funksjon
def tanget(x,c):
    return (f(c) + (andreDerivert(c)*(x-c)))

#newtons metode
#while differense mellom n sted og n-1 steg = 0.00001
#newtowns metode
def tangetSkjeringsPunkt(c):
    #x_(n+1) = x_n -(f(x_n))/f'(x_n)
    intersect = c - f(c)/andreDerivert(c)
    return intersect

#initial x
initi_x = 0.5
newtonTilnerming = tangetSkjeringsPunkt(initi_x)
for forsok in range(2):
    #skjeringspunkt med tangent og x-akse
    ax2.scatter(newtonTilnerming,derivert(newtonTilnerming))
    print(newtonTilnerming)
    newtonTilnerming = tangetSkjeringsPunkt(newtonTilnerming)
    #opptegning av tangent
    ax2.plot(x,tanget(x,newtonTilnerming))





#opptegning av funksjonen
y = f(x)
ax1.plot(x,f(x))
ax1.grid(True)
#opptegning av den deriverte av funksjon
ax2.plot(x,derivert(x))
ax2.grid(True)
#opptegning av den andre deriverte funksjon
ax3.plot(x,andreDerivert(x))
ax3.grid(True)

plt.show()

```
